```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
df.head()
```

# 1. Single Numeric Features

```python
from smarttoolsml.eda_visualisations.single_numeric_fea import boxplot

# single boxplot

series = df["Age"]
boxplot.plot_single_boxplot(df_series=series)
```

```python
# multi boxplot

columns = ["Age", "Fare"]
boxplot.plot_multi_boxplot(df=df, columns=columns)
```

```python
from smarttoolsml.eda_visualisations.single_numeric_fea import histogram

# single histogram

df_series = df["Age"]
histogram.plot_single_hist(df_series=df_series)
```

```python
# multi hist

columns = ["Age", "Fare"]
histogram.plot_multi_hist(df=df, columns=columns)
```

```python
from smarttoolsml.eda_visualisations.single_numeric_fea import kde

# single kde

df_series = df["Age"]
kde.plot_single_kde(df_series=df_series)
```

```python
# multi kde

columns = ["Age", "Fare"]
kde.plot_multi_kde(df=df, columns=columns)
```
